```python
# Split pipe-delimited skills into a long format for analysis
skills_exploded = df.assign(
    skill=df['required_skills'].str.split('|')
).explode('skill')

skills_exploded['skill'] = skills_exploded['skill'].str.strip()

# Top skills by frequency
skills_exploded['skill'].value_counts().head(20)

# Average salary by skill
skills_exploded.groupby('skill')['annual_salary_usd'].mean().sort_values(ascending=False).head(20)

# Skill demand vs salary premium
skills_exploded.groupby('skill').agg(
    avg_salary=('annual_salary_usd', 'mean'),
    count=('skill', 'count'),
    avg_premium=('ai_salary_premium_pct', 'mean')
).sort_values('avg_salary', ascending=False)
```

```python
# Sort by count too — you want skills that are both high-paying AND common enough to matter
skills_exploded.groupby('skill').agg(
    avg_salary=('annual_salary_usd', 'mean'),
    count=('skill', 'count')
).query('count >= 20').sort_values('avg_salary', ascending=False).head(15)

# Also check: top skills by demand_score, not just salary — "hot but not yet highest paid" is a good secondary story
skills_exploded.groupby('skill').agg(
    avg_demand=('demand_score', 'mean'),
    avg_growth=('demand_growth_yoy_pct', 'mean'),
    count=('skill', 'count')
).query('count >= 20').sort_values('avg_growth', ascending=False).head(15)
```
